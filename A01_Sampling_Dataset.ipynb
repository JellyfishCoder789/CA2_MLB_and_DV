##### Dropping comments from before the Hamas attack
 There are only 7 comments in the dataset from before the attack, and so they are removed, as there is not enough data to make assumptions based on these 7 comments.

```python
data_to_drop = df[df['created_date'] == pd.to_datetime("2023-10-06").date()].index

df = df.drop(data_to_drop)
```

```python
sampling_ratio = 0.1  # 10%

reduced_df = df.groupby('created_date').apply(lambda x: x.sample(frac = sampling_ratio)).reset_index(drop = True)
```
